```python
print('Bedroom skewness :',skew(df.Bedrooms, bias=False))
print('Bathrooms skewness :',skew(df.Bathrooms, bias=False))
print('HousePrice skewness :',skew(df.HousePrice, bias=False))
print('SquareFootage skewness :',skew(df.SquareFootage, bias=False))
```

```python
plt.scatter(df.SquareFootage,df.HousePrice)
plt.title('House Prices Vs SquareFootage')
plt.xlabel('House Price')
plt.ylabel('SquareFootage (sft)')
plt.show()
```
